```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurando o estilo dos gráficos para ficar apresentável
sns.set_theme(style="whitegrid")

# Carregando os dados
df = pd.read_csv('../data/spotify-2023.csv', encoding='latin1')

# Visualizando as primeiras 5 linhas
df.head()
```

```python
# Tentando converter a coluna streams para numérico para ver se há erros
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

# Agora vamos ver quantos valores ficaram vazios (NaN) após a conversão
print(f"Valores inválidos encontrados em 'streams': {df['streams'].isna().sum()}")
```

```python
# 1. Removemos as linhas que ficaram com NaN (o valor inválido que encontramos)
df_clean = df.dropna(subset=['streams'])

# 2. Selecionamos as colunas para a nossa análise
features = ['bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%']
df_model = df_clean[features]

# 3. Vamos ver a média dessas características
print(df_model.mean())
```

```python
from sklearn.preprocessing import StandardScaler

# Criamos o escalonador
scaler = StandardScaler()

# Aplicamos o ajuste aos nossos dados
dados_escalonados = scaler.fit_transform(df_model)

# Transformamos de volta em um DataFrame para facilitar a visualização
df_scaled = pd.DataFrame(dados_escalonados, columns=features)

df_scaled.head()
```

```python
from sklearn.cluster import KMeans

# 1. Criamos o modelo com 4 grupos
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

# 2. Treinamos o modelo usando os dados escalonados (df_scaled)
# E já salvamos os rótulos diretamente no nosso DataFrame original
df_clean = df_clean.copy()
df_clean['cluster'] = kmeans.fit_predict(df_scaled)

# 3. Verificamos se deu certo
print("Grupos criados com sucesso!")
print(df_clean['cluster'].value_counts())
```

```python
# Calculando a média de cada característica por grupo
resumo_clusters = df_clean.groupby('cluster')[features].mean()
print(resumo_clusters)
```

```python
# Filtrando apenas as músicas que caíram no grupo mais acústico/calmo
musicas_calmas = df_clean[df_clean['cluster'] == 2]

# Exibindo as primeiras 10 músicas desse grupo
print("Exemplos de músicas no Cluster 2 (Mais calmas/acústicas):")
print(musicas_calmas[['track_name', 'artist(s)_name']].head(10))
```

```python
df_clean.to_csv('spotify_com_clusters.csv', index=False, encoding='utf-8')
print("✅ Arquivo 'spotify_com_clusters.csv' gerado com sucesso!")
```

```python
from sklearn.decomposition import PCA

# 1. Reduzimos as 5 dimensões para 2 para conseguirmos plotar
pca = PCA(n_components=2)
componentes = pca.fit_transform(df_scaled)

# 2. Criamos um DataFrame temporário para o gráfico
df_pca = pd.DataFrame(data=componentes, columns=['PC1', 'PC2'])
df_pca['cluster'] = df_clean['cluster'].values

# 3. Geramos o gráfico de dispersão
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='cluster', palette='viridis', s=60)
plt.title('Mapa Espacial dos Clusters de Músicas (PCA)')
plt.show()
```
